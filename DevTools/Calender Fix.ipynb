```python
slash = '/'
path = os.getcwd()
export_loc = path + slash + "Exports"
```

```python
cal = pd.read_csv(path + slash + "cal2.csv", infer_datetime_format=True)
cal.columns= cal.columns.str.lower()
cal["date"] = cal["date"].str[:-10]
cal["date"] = pd.to_datetime(cal["date"], dayfirst = True)
cal.sort_values(by="date",inplace=True)
cal=cal.dropna()
cal.reset_index(inplace=True,drop=True)
cal
```

```python
n_split = cal["title"].str.split(',', expand=True)
n_split.columns = ["surname","name","package"]
n_split
```

```python
cal = pd.concat([n_split, cal], axis=1)
cal = cal.drop(columns='title')
cal = cal.reindex(columns=["date","surname","name","package"])
cal
```

```python
cal['package'] = cal['package'].replace({'1': 'single',
                                         '5': 'five',
                                         '8': 'eight',
                                         '10': 'ten',
                                         '12': 'twelve',
                                         '20': 'twenty'})
cal
```

```python
cal['payment'] = np.arange(len(cal))

cal["payment"][cal['package'] == "single"] = 60
cal["payment"][cal['package'] == "five"] = 287.5
cal["payment"][cal['package'] == "eight"] = 440
cal["payment"][cal['package'] == "ten"] = 525
cal["payment"][cal['package'] == "twelve"] = 600
cal["payment"][cal['package'] == "twenty"] = 900
cal
```

```python
cal["payment type"] = "bank"
cal["payment type"][cal['package'] == "single"] = "cash"
cal
```

```python
cal.to_csv(path + slash + "calex.csv",index=False)
```
